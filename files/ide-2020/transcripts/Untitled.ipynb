```python
from datetime import datetime
```

```python
my_dict = {}

with open('transcript1-1.txt') as txtfile:
    cnt = 0
    for line in txtfile:
        if (cnt%3)==0:
            tm = line
        if (cnt%3)==1:
            txt = line
        if (cnt%3)==2:
            my_dict[tm] = txt
        cnt += 1
            
```
